```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("EmployeeAttrition.csv")
```

```python
df.columns
```

```python
corr = df.corr()
plt.subplots(figsize=(25,15))
sns.heatmap(corr, annot=True)
```

```python
#split dependent/independent
x = df.drop(['Attrition','EmployeeCount','EmployeeNumber','EducationField','Over18','StandardHours'],axis=1)
y = df.Attrition
```

```python
x.head()
```

```python
x.shape,y.shape
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
x.BusinessTravel = le.fit_transform(x.BusinessTravel)
x.Department = le.fit_transform(x.Department)
x.Gender = le.fit_transform(x.Gender)
x.JobRole = le.fit_transform(x.JobRole)
x.MaritalStatus = le.fit_transform(x.MaritalStatus)
x.OverTime = le.fit_transform(x.OverTime)
```

```python
y
```

```python
x.head()
```

```python
#Splitting Data into Train and Test.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

##Logistic Regression

```python
#Model Building
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=10000)
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
pred
```

Evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
print("Accuracy Score")
print(accuracy_score(y_test,pred))
print("\nConfusion Matrix")
print(confusion_matrix(y_test,pred))
print("\nCross Tab")
print(pd.crosstab(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
probability = lr.predict_proba(x_test)[:,1]
fpr,tpr,threshsholds = roc_curve(y_test,probability)
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```

##Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
pred = dt.predict(x_test)
pred
```

```python
from sklearn import tree
plt.figure(figsize=(25,15))
tree.plot_tree(dt,filled=True)
```

```python
from sklearn.model_selection import GridSearchCV
#prepruning
parameters = {
    'criterion':['gini','entropy'],
    'splitter':['best','random'],
    'max_depth':list(range(1,10)),
    'max_features':['sqrt','log2']  #`max_features='auto'` has been deprecated in 1.1
}
```

```python
gscv= GridSearchCV(estimator=dt, param_grid=parameters, cv=5, scoring = 'accuracy')
```

```python
gscv.fit(x_train, y_train)
```

```python
gscv.best_params_
```

```python
dt_cv = DecisionTreeClassifier(criterion= 'gini',
 max_depth=4,
 max_features='sqrt',
 splitter='best'
)

dt_cv.fit(x_train,y_train)
```

Evaluation

```python
pred = dt_cv.predict(x_test)
print("Accuracy Score")
print(accuracy_score(y_test,pred))
print("\nConfusion Matrix")
print(confusion_matrix(y_test,pred))
print("\nCross Tab")
print(pd.crosstab(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
probability = dt_cv.predict_proba(x_test)[:,1]
fpr,tpr,threshsholds = roc_curve(y_test,probability)
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```

##Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(max_features='sqrt')
forest_params = [{'max_depth': list(range(10,15)), 'max_features': list(range(0,19))}]
rfc_cv = GridSearchCV(rfc, forest_params, cv=10, scoring='accuracy')
```

```python
rfc_cv.fit(x_train, y_train)
```

```python
rfc_cv.best_params_
```

```python
rfc_cv = RandomForestClassifier(
 max_depth=11,
 max_features=16,
)

rfc_cv.fit(x_train,y_train)
```
